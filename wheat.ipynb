##### The dataset contains information on wheat cultivation in various districts of India. The data includes features such as district code, year, state code, state name, district name, wheat area in 1000 hectares, wheat production in 1000 tons, and wheat yield in kg per hectare

```python
# Load  the data

dataset = pd.read_csv(r"D:\NEHA\EDA\Wheat.csv")
df = pd.DataFrame(dataset)
df
```

##### After reading the data we transform it into a dataframe. Here we can see that the data consist of 16146 rows and 8 different columns.

# Sorting and  Comparison Operations

```python
# Calculate the mean wheat production of each states 

df.groupby("state_name")["wheat_production"].mean()
```

```python
# Calculate the maximum wheat production of each states 

df.groupby("state_name")["wheat_production"].max()
```

```python
max_prod_state = df.loc[df['wheat_production'].idxmax(), 'state_name']
max_prod_value = df['wheat_production'].max()

print('The state with maximum wheat production is', max_prod_state, 'with a production of',max_prod_value)
```

```python
# Sum of wheat production in each states

prod_sum = df.groupby("state_name")["wheat_production"].sum()
prod_sum
```

```python
#State with zero wheat production

prod_sum_zero = prod_sum[prod_sum == 0]
prod_sum_zero
```

```python
# To find the average wheat production and wheat area of different states 

df.groupby(["year","state_name"]).agg(
    avg_wheatproduction =("wheat_production","mean"),
    avg_wheatarea=("wheat_area","mean"))
```

```python
# To find the maximum wheat production and wheat area of different states 

df.groupby(["year","state_name"]).agg(
    max_wheatproduction =("wheat_production","max"),
    max_wheatarea=("wheat_area","max"))
```

```python
# To find the min wheat production and wheat area of different states 

df.groupby(["year","state_name"]).agg(
    min_wheatproduction =("wheat_production","min"),
    min_wheatarea =("wheat_area","min"))
```

# Graphical Representations

### Finding correlation between variables using heatmap

```python
# Heatmap

plt.figure(figsize=(8,5))
sns.heatmap(df.corr(),annot=True,linewidths=0.5,cmap="Blues")
```

```python
# Scatter plot

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.scatterplot(x='wheat_area',y='wheat_production',data=df, color='indigo')
plt.xlabel('Wheat Area')
plt.ylabel('Wheat Production')
plt.title('Wheat Area vs Wheat Production')


plt.subplot(1,2,2)
sns.scatterplot(x='wheat_production',y='wheat_yield',data=df, color='orange')
plt.xlabel('Wheat Area')
plt.ylabel('Wheat Yield')
plt.title('Wheat Area vs Wheat Yield')
plt.tight_layout()
plt.show()
```

```python
# Bar chart showing the wheat production over different years

plt.figure(figsize=(8,5))
x = df['year']
y = df['wheat_production']
plt.xlabel('Year')
plt.ylabel('Wheat Production')
plt.title('Wheat Production in each Year')
plt.bar(x,y,color="green")
plt.show()
```

```python
# Histogram of wheat yield

plt.figure(figsize=(8,5))
df['wheat_yield'].plot(kind='hist', bins=20, title='wheat_yield',color="purple",edgecolor='k')
plt.show()
```

```python
# Bar chart showing the states with wheat production greater than zero

plt.figure(figsize=(8,5))
df_prod = df[df['wheat_production']>0]
prod_state = df_prod['state_name'].value_counts()
prod_state.plot(kind='bar',color='hotpink')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot showing wheat production of different states over several years

plt.figure(figsize=(15,10))

# Plot for 1970
plt.subplot(2,3,1)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 1970])
plt.xticks(rotation=90)
plt.title('Wheat Production in 1970')

# Plot for 1980
plt.subplot(2,3,2)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 1980])
plt.xticks(rotation=90)
plt.title('Wheat Production in 1980')

# Plot for 1990
plt.subplot(2,3,3)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 1990])
plt.xticks(rotation=90)
plt.title('Wheat Production in 190')

# Plot for 2000
plt.subplot(2,3,4)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 2000])
plt.xticks(rotation=90)
plt.title('Wheat Production in 2000')

# Plot for 2010
plt.subplot(2,3,5)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 2010])
plt.xticks(rotation=90)
plt.title('Wheat Production in 2010')

# Plot for 2017
plt.subplot(2,3,6)
sns.lineplot(x='state_name',y='wheat_production',data=df[df['year'] == 2017])
plt.xticks(rotation=90)
plt.title('Wheat Production in 2017')

plt.tight_layout()
plt.show()
```

```python
# Plot showing wheat area of different states over several years

# Filter data for each year
data_1970 = df[df['year'] == 1970]
data_1980 = df[df['year'] == 1980]
data_1990 = df[df['year'] == 1990]
data_2000 = df[df['year'] == 2000]
data_2010 = df[df['year'] == 2010]
data_2017 = df[df['year'] == 2017]

plt.figure(figsize=(15, 10))

# Plot for 1970
plt.subplot(2, 3, 1)
plt.bar(data_1970['state_name'], data_1970['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 1970')

# Plot for 1980
plt.subplot(2, 3, 2)
plt.bar(data_1980['state_name'], data_1980['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 1980')

# Plot for 1990
plt.subplot(2, 3, 3)
plt.bar(data_1990['state_name'], data_1990['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 1990')

# Plot for 2000
plt.subplot(2, 3, 4)
plt.bar(data_2000['state_name'], data_2000['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 2000')

# Plot for 2010
plt.subplot(2, 3, 5)
plt.bar(data_2010['state_name'], data_2010['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 2010')

# Plot for 2017
plt.subplot(2, 3, 6)
plt.bar(data_2017['state_name'], data_2017['wheat_area'],color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Wheat Area in 2017')

plt.tight_layout()
plt.show()
```
